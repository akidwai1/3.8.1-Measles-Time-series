a. Edit your population genetics simulator to output the allele frequency of A for each generation, so that you could produce a graph like those in slide 16 from today's powerpoint.  First, simply print this information to screen. Next, save the allele frequencies in a list and output them as a .csv file. Run the code with A = 0.5 starting frequency and N = 10 (pop) and N = 100 population sizes.

b. Submit your code as a .py file and your output files as a .csv

```python
import random
import csv

# Function to simulate a single generation of population genetics
def simulate_generation(A, N):
    # Create a population with given initial allele frequency
    population = ['A' if random.random() < A else 'a' for _ in range(N)]
    
    # Calculate the allele frequency of A in the population
    freq_A = population.count('A') / N
    
    return freq_A

# Initialize parameters
A = 0.5
population_sizes = [10, 100]

# Output allele frequencies to screen and save to CSV file
for N in population_sizes:
    allele_frequencies = []
    for _ in range(10):  # 10 generations
        freq_A = simulate_generation(A, N)
        allele_frequencies.append(freq_A)
        
        # Output allele frequency to screen
        print(f"Population size: {N}, Generation: {_+1}, Allele frequency of A: {freq_A}")
    
    # Save allele frequencies to a CSV file
    with open(f'allele_frequencies_N{N}.csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(allele_frequencies)
```

```python
import random
import matplotlib.pyplot as plt

def population_genetics_simulator(starting_freq_A, population_size):
    generations = 10
    freq_A = [starting_freq_A]

    for gen in range(1, generations + 1):
        new_freq_A = 0
        for _ in range(population_size):
            p = freq_A[-1]  # current frequency of A
            # Simulate genetic drift
            new_p = random.random()
            if new_p < p:
                new_freq_A += 1

        freq_A.append(new_freq_A / population_size)

    return freq_A

# Simulate with A = 0.5 starting frequency and N = 10 population size
freq_A_pop_10 = population_genetics_simulator(0.5, 10)

# Simulate with A = 0.5 starting frequency and N = 100 population size
freq_A_pop_100 = population_genetics_simulator(0.5, 100)

# Plot the graphs
plt.plot(range(11), freq_A_pop_10, label='Population Size = 10')
plt.plot(range(11), freq_A_pop_100, label='Population Size = 100')
plt.xlabel('Generation')
plt.ylabel('Frequency of A allele')
plt.title('Population Genetics Simulation')
plt.legend()
plt.show()

```
